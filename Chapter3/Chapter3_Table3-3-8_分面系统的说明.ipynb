```python
from plotnine import *
from plotnine.data import mpg
```

```python
mpg.head()
```

```python
mpg.fl.value_counts()
```

```python
t = (ggplot(mpg, aes(x='cty', y='hwy', fill='fl'))
     + geom_point(size=3, stroke=0.3, alpha=0.8, show_legend=False)
     + scale_fill_hue(s=0.9, l=0.65, h=0.0417, color_space='husl')
     + theme(figure_size=(4, 3), dpi=100)
    )
print(t)
```

```python
mpg['fl'].value_counts()
```

```python
t1 = (t + facet_grid('. ~ fl') # .
      + theme_matplotlib()
      + theme(text=element_text(color='black', size=10), 
             strip_text=element_text(color='black', size=15), # 
             strip_background=element_rect(color='black', fill='lightgray'), #
             aspect_ratio=1.3,
             dpi=100,
             figure_size=(10, 5)) # figure_size,facet
     )
print(t1)
```

```python
t2 = (t + facet_grid('year ~ .')
      + theme_xkcd()
      + theme(text=element_text(color='black', size=12),
             strip_text=element_text(color='black', size=12), # themetextelement_text
             strip_background=element_rect(color='black'),
             aspect_ratio=0.5,
             dpi=100,
             figure_size=(4, 4)) 
     )
print(t2)
```

```python
t3 = (t + facet_grid('year ~ fl')
      + theme_matplotlib()
      + theme(text=element_text(color='black', size=12),
             strip_text=element_text(color='black', size=15), # themetextelement_text
             strip_background=element_rect(color='black'),
             aspect_ratio=1,
             dpi=100,
             figure_size=(10, 8)) 
     )
print(t3)
```

```python
t4 = (t + facet_wrap('~ fl') # t1facdet_gridfacet_wrap
      + theme_matplotlib()
      + theme(text=element_text(color='black', size=12),
             strip_text=element_text(color='black', size=15),
             strip_background=element_rect(color='black'),
             aspect_ratio=1,
             dpi=100,
             figure_size=(8, 5)) 
     )
print(t4)
```

```python
t5 = (t + facet_grid('drv ~ fl', scales='fixed') 
      # scales:xy, fixed:subgroup,xy;free:x,y;free_x:xy
      + theme_matplotlib()
      + theme(text=element_text(color='black', size=12),
             strip_text=element_text(color='black', size=15),
             strip_background=element_rect(color='black'),
             aspect_ratio=1,
             dpi=100,
             figure_size=(8, 5)) 
     )
print(t5)
```

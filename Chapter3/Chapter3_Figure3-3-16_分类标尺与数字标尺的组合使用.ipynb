```python
p1 = (ggplot(df, aes(x='variable', y='value'))
      + geom_col(width=0.5, fill='orange')
      + theme_xkcd()
      + theme(figure_size=(4, 3))
     )
print(p1)
```

```python
p2 = (ggplot(df, aes(x='variable', y='value'))
      + geom_col(width=0.5)
      + coord_flip()
      + theme_matplotlib()
      + theme(figure_size=(5, 3))
     )
print(p2)
```

(p + scale_fill_distiller(type='div', palette='RdYlBu')): 使用双渐变色系'RdYlBu'颜色主题方案

(p + scale_fill_cmap(name='viridis')): 使用'viridis'颜色主题方案

(p + scale_fill_gradient2(low='#00A08A', mid='white', high='#FF0000', midpoint=np.mean(df.age))): 自定义连续的颜色条，np.mean(df.age))表示age均值对应中间色'white'

(p + scale_fill_gradientn(colors= ("#82C143","white","#CB1B81"))): 使用Hex颜色码自定义填充颜色

```python
p1 = (ggplot(df, aes(x='SOD', y='tau', fill='age'))
      + geom_point(shape='o', size=3, color='black', stroke=0.25)
      + scale_fill_distiller(type='div', palette='RdYlBu')
      + guides(fill=guide_colorbar(barheight=8, barwidth=5)) # 设置图例的高度和宽度
      + theme(text=element_text(color='black', size=12),
             aspect_ratio=1.15,              
             dpi=100,
             legend_position=(0.8, 0.25),
             legend_direction='vertical',
             legend_background=element_blank(),
             figure_size=(4, 4))
     )
print(p1)
```

```python
p2 = (ggplot(df, aes(x='SOD', y='tau', fill='age'))
      + geom_point(shape='o', size=3, color='black', stroke=0.25)
      + scale_fill_cmap(cmap_name='viridis') # 可以随意输入string, 然后根据报错选择合适的colormap; colormaps: https://matplotlib.org/2.0.2/users/colormaps.html
      + guides(fill=guide_colorbar(barheight=8, barwidth=6))
      + theme(text=element_text(color='black', size=10),
             aspect_ratio=1.15,              
             dpi=100,
             legend_position=(0.8, 0.25),
             legend_direction='vertical',
             legend_background=element_blank(),
             figure_size=(4, 4))
     )
print(p2)
```

```python
p3 = (ggplot(df, aes(x='SOD', y='tau', fill='age'))
      + geom_point(shape='o', size=3, stroke=0.25, color='black')
      + scale_fill_gradient2(low="#00A08A",mid="white",high="#FF0000",midpoint = np.mean(df.age)) #np.mean(df.age))表示age均值对应中间色'white'
      + guides(fill=guide_colorbar(barheight=7, barwidth=6))
      + theme_matplotlib()
      + theme(text=element_text(color='black', size=10),
              aspect_ratio=1.15,
              dpi=100,
              figure_size=(4, 4),
              legend_position=(0.8, 0.25),
              legend_direction='vertical',
              legend_background=element_blank())
     )
print(p3)
```

```python
p4 = (ggplot(df, aes(x='SOD', y='tau', fill='age'))
      + geom_point(shape='o', size=3, stroke=0.25, color='black')
      + scale_fill_gradientn(colors=("#31a354","white","#c51b8a"))
      + guides(fill=guide_colorbar(barheight=6.5, barwidth=6))
      + theme_xkcd()
      + theme(text=element_text(color='black', size=10),
              aspect_ratio=1.15,
              dpi=100,
              figure_size=(4, 4),
              legend_position=(0.8, 0.25),
              legend_direction='vertical',
              legend_background=element_blank())
     )
print(p4)
```

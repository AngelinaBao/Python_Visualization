* (p + scale_fill_discrete()): plotnine默认配色方案<br>
* (p + scale_fill_brewer(type='qualitative', palette='Set1')): 使用Set1的多色系颜色主题方案<br>
* (p + scale_fill_hue(s=1, l=0.65, h=0.0417, color_space='husl')): 使用HSLuv的离散型颜色主题方案<br>
* (p + scale_fill_manual(values=["#E7298A","#66A61E","#E6AB02"])): 使用Hex颜色码自定义填充颜色

**不同的scale_fill_xx只是使用不同的颜色模式。**

```python
p1 = (ggplot(df, aes(x='SOD', y='tau', fill='Class'))
      + geom_point(shape='o', size=2, color='black', stroke=0.25)
      + scale_fill_discrete()
      + theme(text=element_text(color='black', size=10), 
              legend_position=(0.75, 0.25),
              legend_direction='vertical',
              legend_background=element_blank(),
              aspect_ratio=1.05,
              dpi=100, 
              figure_size=(4, 4))
     )
print(p1)
```

**scale_fill_brewer(type='seq', palette=1, direction=1, **kwargs)**

type : str in ``['seq', 'div', 'qual']`` <br>
>Type of data. Sequential, diverging or qualitative
    
* type: diverging<br>
>palettes: BrBG, PiYG, PRGn, PuOr, RdBu, RdGy, RdYlBu, RdYlGn, Spectral

* type: qualitative<br>
>Accent, Dark2, Paired, Pastel1, Pastel2, Set1, Set2, Set3

* type: sequential<br>
>Blues, BuGn, BuPu, GnBu, Greens, Greys, Oranges, OrRd, PuBu, PuBuGn, PuRd, Purples, RdPu, Reds, YlGn, YlGnBu, YlOrBr, YlOrRd

direction: int in ``[-1, 1]``<br>
>Sets the order of colors in the scale. If 1, the default, colors are as output by mizani.palettes.brewer_pal(). If -1, the order of colors is reversed.

```python
p2 = (ggplot(df, aes(x='SOD', y='tau', fill='Class'))
      + geom_point(shape='o', size=2, color='black', stroke=0.25)
      + scale_fill_brewer(type='qual', palette='Set1') # 颜色比p1更鲜艳
      + theme(text=element_text(color='black', size=10), 
              legend_position=(0.75, 0.25),
              legend_direction='vertical',
              legend_background=element_blank(),
              aspect_ratio=1.05,
              dpi=100, 
              figure_size=(4, 4))
     )
print(p2)
```

```python
p3 = (ggplot(df, aes(x='SOD', y='tau', fill='Class'))
      + geom_point(shape='o', size=2, color='black', stroke=0.25)
      + scale_fill_hue(s = 1, l = 0.65, h=0.0417,color_space='husl')
      + theme(text=element_text(color='black', size=10), 
              legend_position=(0.75, 0.25),
              legend_direction='vertical',
              legend_background=element_blank(),
              aspect_ratio=1.05,
              dpi=100, 
              figure_size=(4, 4))
     )
print(p3)
```

```python
p4 = (ggplot(df, aes(x='SOD', y='tau', fill='Class'))
      + geom_point(shape='o', size=2, color='black', stroke=0.25)
      + scale_fill_manual(values=["#f03b20","#2b8cbe","#fed98e"]) # 颜色Hex查询 colorbrewer: https://colorbrewer2.org/#type=sequential&scheme=BuGn&n=3
      + theme_matplotlib()
      + theme(text=element_text(color='black', size=10), 
              legend_position=(0.75, 0.25),
              legend_direction='vertical',
              legend_background=element_blank(),
              aspect_ratio=1.05,
              dpi=100, 
              figure_size=(4, 4))
     )
print(p4)
```

```python
p5 = (ggplot(df, aes(x='SOD', y='tau', fill='Class'))
      + geom_point(shape='o', size=2, color='black', stroke=0.25)
      + scale_fill_manual(values=["#f03b20","#2b8cbe","#fed98e"])
      + geom_smooth(aes(color='Class'), size=.8, se=False, method='lm')
      + scale_color_manual(values=["#f03b20","#2b8cbe","#fed98e"])
      + theme_matplotlib()
      + theme(text=element_text(color='black', size=10), 
              legend_position=(0.75, 0.25),
              legend_direction='vertical',
              legend_background=element_blank(),
              aspect_ratio=1.05,
              dpi=100, 
              figure_size=(4, 4))
     )
print(p5)
```

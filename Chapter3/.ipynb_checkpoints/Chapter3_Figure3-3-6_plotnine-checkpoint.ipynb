### plotnine不同美学参数映射效果

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from plotnine import *

mpl.rcParams['font.family'] = 'Times New Roman'

%matplotlib inline
%config InlineBackend.figure_format='svg'
```

```python
df = pd.read_csv(r'C:\Users\1000257489\Documents\2022\Angelina\Study\Beautiful Visualization with Python\Beautiful-Visualization-with-python-master\第3章_数据可视化基础\Facet_Data.csv')
df.shape
```

#### age映射到点的大小

```python
'''
for geom_point: 
shape: point shape, 
color: point edge color(if fill=None(default), the filled color will also be the color you set)
storke: point edge width
The aspect_ratio takes precedence over the figure_size, and it modifies the height. The effective width and height are:
width = figure_size[0]
height = figure_size[0] * aspect_ratio

如果fill is not None, color in geom_point will be edge color of the point, filled color will be different per fill variable;
else, color will be filled and edge color at same time
'''
p1 = (ggplot(df, aes(x='SOD', y='tau', size='age'))
     + geom_point(shape='o', color='black', fill='#336A97', stroke=0.25, alpha=0.6)
     + theme(text=element_text(size=12, color='black'), aspect_ratio=1, dpi=100, figure_size=(4, 4))) 
print(p1)
```

#### age映射到点的大小和填充颜色

```python
p2 = (ggplot(df, aes(x='SOD', y='tau', size='age', fill='age'))
     + geom_point(shape='o', color='black', stroke=0.25, alpha=0.8)
     + theme(text = element_text(size=12, color='black'), aspect_ratio=1, dpi=100, figure_size=(4, 4))
     )
print(p2)
```

#### Class映射到散点的填充颜色

```python
p3 = (ggplot(df, aes(x='SOD', y='tau', fill='Class'))
      + geom_point(shape='o', color='black', stroke=0.25, size=3)
      + theme(text = element_text(color='black', size=12), aspect_ratio=1, dpi=100, figure_size=(4, 4))
     )
print(p3)
```

#### age和Class分别映射到散点的大小(size)和填充颜色(fill)

```python
p4 = (ggplot(df, aes(x='SOD', y='tau', size='age', fill='Class'))
      + geom_point(shape='o', color='black', stroke=0.25, alpha=0.8)
      + theme(text = element_text(color='black', size=12), 
              dpi=100, 
              aspect_ratio=1, 
              figure_size=(4, 4))
     )
print(p4)
```

### 度量调整

#### 散点大小(size)的度量调整

```python
p1 = (ggplot(df, aes(x='SOD', y='tau', size='age')) 
      + geom_point(shape='o', color='black', fill='red', stroke=0.25, alpha=0.8)
      + scale_size(range=(1, 8))
      + theme(text = element_text(color='black', size=12), aspect_ratio=1, dpi=100, figure_size=(4, 4))
     )
print(p1)
```

#### 散点大小(size)和填充颜色(fill)的调整

```python
p2 = (ggplot(df, aes(x='SOD', y='tau', fill='age', size='age'))
      + geom_point(shape='o', color='black', stroke=0.25, alpha=0.8) 
      + scale_size(range=(1, 8))
      + scale_fill_distiller(type='seq', palette='Reds') # 这个语句有问题,图例不会更新为新的颜色;加上guides(fill=guide_legend(title='age'))就可以了
      + theme(text = element_text(color='black', size=12), aspect_ratio=1, dpi=100, figure_size=(4, 4))
      + guides(fill=guide_legend(title='age'))
     )
print(p2)
```

#### 点颜色填充fill与形状shape的度量调整

```python
p3=(ggplot(df, aes(x='SOD',y='tau',fill='Class',shape='Class')) + 
  geom_point(size=3,colour="black",stroke=0.25)+
  scale_fill_manual(values=("#36BED9","#FF0000","#FBAD01"))+
   scale_shape_manual(values=('o','s','D'))+
   #scale_fill_hue(s = 0.90, l = 0.65, h=0.0417,color_space='husl')+
             theme(text=element_text(size=12,colour = "black"),
        aspect_ratio =1,
        dpi=100,
       figure_size=(4,4))) #shape=21,color="black",fill="red",size=3,stroke=0.1
print(p3)
```

#### 点大小size和颜色fill的度量调整

```python
p4 = (ggplot(df, aes(x='SOD', y='tau', size='age', fill='Class'))
      + geom_point(shape='o', color='black', stroke=0.25, alpha=0.8)
      + scale_fill_manual(values=["#36BED9","#FF0000","#FBAD01"])
      + scale_size(range=(1, 8))
      + theme(text = element_text(color='black', size=12), aspect_ratio=1, dpi=100, figure_size=(4, 4))
     )
print(p4)
```

```python
# playful code
p1 = (ggplot(df, aes(x='SOD', y='tau'))
     + geom_point(shape='.', color='gray', size=2, alpha=0.8)
     + geom_smooth(method='lm', size=0.5)
     + theme(text=element_text(size=12, color='black'), aspect_ratio=1, dpi=100, figure_size=(4, 4))) 
print(p1)
```

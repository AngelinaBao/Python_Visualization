```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='svg'

mpl.rcParams['font.family'] = 'SimHei'
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.sans-serif'] = 'SimHei'
plt.style.use('default')
```

```python
df = pd.read_csv(r'C:\Users\1000257489\Documents\2022\Angelina\Study\Beautiful Visualization with Python\Beautiful-Visualization-with-python-master\第3章_数据可视化基础\MappingAnalysis_Data.csv')
df.shape
```

```python
df.head()
```

```python
df['variable'].value_counts()
```

```python
df1 = df.loc[df['variable']=='0%(Control)']
df2 = df.loc[df['variable']=='1%']
df3 = df.loc[df['variable']=='5%']
df4 = df.loc[df['variable']=='15%']
```

```python
group = df['variable'].unique().tolist()
colors = ['#e41a1c','#377eb8','#4daf4a','#984ea3']
markers = ['o','s','H','D']    
labels = group
```

```python
# figure (a)
fig = plt.figure(figsize=(4, 3), dpi=100)
for i in range(4):
    temp_df = df.loc[df['variable']==group[i]]
    plt.plot(temp_df.Time, temp_df.value)
```

```python
# figure(b)
fig = plt.figure(figsize=(4, 3), dpi=100)
for i in range(4):
    plt.plot(df.loc[df['variable']==group[i]].Time, df.loc[df['variable']==group[i]].value, 
             marker=markers[i], mfc=colors[i], ms=6, mew=0.5, color='k', ls='-', label=labels[i])
plt.legend(loc='best', frameon=False);
```

```python
# figure(c)
fig = plt.figure(figsize=(6, 4), dpi=100)

plt.plot(df1.Time, df1.value, marker=markers[0], ms=6, mfc=colors[0], mew=0.5, color='k', ls='-', lw=2, label=labels[0])
plt.plot(df2.Time, df2.value, marker=markers[1], ms=6, mfc=colors[1], mew=0.5, color='k', ls='-', lw=2, label=labels[1])
plt.plot(df3.Time, df3.value, marker=markers[2], ms=6, mfc=colors[2], mew=0.5, color='k', ls='-', lw=2, label=labels[2])
plt.plot(df4.Time, df4.value, marker=markers[3], ms=6, mfc=colors[3], mew=0.5, color='k', ls='-', lw=2, label=labels[3])

plt.xlabel('Time(d)', fontsize=12)
plt.ylabel('value', fontsize=12)
plt.xticks(np.linspace(0, 20, 11), fontsize=10)
plt.yticks(np.linspace(0, 90, 10), fontsize=10)
plt.legend(loc='best', frameon=True)
ax = plt.gca()

from enum import Enum

class SpinesEnum(Enum):
    right='right'
    top='top'
    
# ax.spines[Spines.right.value].set_color('r') # 设置边框颜色
    
class Spines:
    right='right'

# spines = Spines()

ax.spines[Spines.right].set_color('r') # 设置边框颜色
ax.spines['top'].set_color('b')
```

### matplotlib多幅子图和坐标系的添加方法

```python
plt.figure()
plt.subplot(221, fc='y')
plt.subplot(222, fc='b')
plt.subplot(212, fc='g')
plt.show()
```

```python
figure = plt.figure()
ax1 = figure.add_subplot(221, fc='y')
ax2 = figure.add_subplot(222, fc='b')
ax3 = figure.add_subplot(212, fc='g')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))
plt.show()
```

```python
plt.subplot2grid((2, 3), (0, 0), colspan=2, facecolor='y')
plt.subplot2grid((2, 3), (0, 2), facecolor='b')
plt.subplot2grid((2, 3), (1, 0), colspan=3, facecolor='g')
plt.show()
```

```python
import matplotlib.gridspec as gridspec
G = gridspec.GridSpec(2, 3)
ax1 = plt.subplot(G[0, 0:2], fc='y')
ax2 = plt.subplot(G[0, 2], fc='c')
ax3 = plt.subplot(G[1, :], fc='m')
plt.show()
```

```python
plt.axes([0.1, 0.1, .8, .8], fc='m')
plt.axes([.2, .2, .3, .3], fc='y')
plt.show()
```

```python
plt.axes([.1, .1, .5, .5], fc='y')
plt.axes([.2, .2, .5, .5], fc='r')
plt.axes([.3, .3, .5, .5], fc='b')
plt.axes([.4, .4, .5, .5], fc='g')
plt.show()
```

### 图3-1-4 matplotlib子图绘制案例

```python
fig = plt.figure(figsize=(7, 3), dpi=100)
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(df1.Time, df1.value, marker=markers[0], ms=6, mfc=colors[0], mew=0.5, color='k', ls='-', lw=2, label=labels[0])
ax1.plot(df2.Time, df2.value, marker=markers[1], ms=6, mfc=colors[1], mew=0.5, color='k', ls='-', lw=2, label=labels[1])
ax1.set_xlabel('Time(d)', fontsize=12)
ax1.set_ylabel('value', fontsize=12)
ax1.set_xticks(np.linspace(0, 20, 11))
ax1.set_yticks(np.linspace(0, 90, 10))
ax1.legend(loc='upper left', frameon=False)

ax2 = fig.add_subplot(1, 2, 2, sharey=ax1)
ax2.plot(df1.Time, df1.value, marker=markers[0], ms=6, mfc=colors[0], mew=0.5, color='k', ls='-', lw=2, label=labels[0])
ax2.plot(df3.Time, df3.value, marker=markers[2], ms=6, mfc=colors[2], mew=0.5, color='k', ls='-', lw=2, label=labels[2])
ax2.set_xlabel('Time(d)', fontsize=12)
ax2.set_xticks(np.linspace(0, 20, 11))
ax2.legend(loc='upper left', frameon=False)

ax1.spines['right'].set_color('none')
ax2.spines['right'].set_color('none')
ax1.spines['top'].set_color('none')
ax2.spines['top'].set_color('none')

plt.suptitle('Different Percents Trend Plot')
plt.subplots_adjust(wspace=0.2)
plt.show()
```

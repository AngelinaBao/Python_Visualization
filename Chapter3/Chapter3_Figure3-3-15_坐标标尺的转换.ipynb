```python
import pandas as pd
import numpy as np
from plotnine import *
import matplotlib

matplotlib.rcParams['font.family'] = 'Times New Roman'
```

```python
df=pd.read_csv(r"C:\Users\1000257489\Documents\2022\Angelina\Study\Beautiful Visualization with Python\Beautiful-Visualization-with-python-master\第3章_数据可视化基础\logarithmic_scale.csv")
df_melt=pd.melt(df,id_vars='VIN(V)',var_name='Class',value_name='value')
```

```python
p1 = (ggplot(df_melt, aes(x='VIN(V)', y='value', color='Class'))
      + geom_line(size=1)
      + scale_x_continuous(breaks=np.arange(-20, 21, 5), limits=(-20, 20))
      + scale_y_continuous(breaks=np.arange(0, 1.8, 0.5), limits=(0, 1.7))
      + theme_classic()
      + theme(text=element_text(color='black', size=10), dpi=100, figure_size=(5, 5),
              panel_grid_major=element_line(color="#C7C7C7",linetype ='--'),
              legend_position=(0.8, 0.8), legend_direction='vertical', legend_background=element_blank())
     )
print(p1)
```

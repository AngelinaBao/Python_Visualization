### 使用seaborn绘制带标记的曲线图

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='svg'

mpl.rcParams['font.family'] = 'SimHei'
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.sans-serif'] = 'SimHei'

import seaborn as sns
sns.set_palette('Set1') #设置绘图的颜色主题
sns.set_style('ticks') #设置绘图的图表风格
sns.set_context(rc={'axes.labelsize': 15, 'legend.fontsize':10, 
                    'xtick.labelsize': 13,'ytick.labelsize': 13}) #设置元素的缩放比例
```

```python
df = pd.read_csv(r'C:\Users\1000257489\Documents\2022\Angelina\Study\Beautiful Visualization with Python\Beautiful-Visualization-with-python-master\第3章_数据可视化基础\MappingAnalysis_Data.csv')
df.shape
```

```python
df.head()
```

```python
# figure(a)
fig = plt.figure(figsize=(6, 4), dpi=100)
sns_line = sns.lineplot(x='Time', y='value', hue='variable', data=df, markers=True, dashes=False)
```

```python
# figure(b)
markers = ['o', 's', 'H', 'D']
fig = plt.figure(figsize=(6, 4), dpi=100)
sns_line = sns.lineplot(x='Time', y='value', hue='variable', style='variable', data=df, 
                        markers=markers, ms=6, dashes=False, mec='k', mew=0.5, lw=2)
```

```python
# figure(c)
markers = ['o', 's', 'H', 'D']
fig = plt.figure(figsize=(6, 4), dpi=100)
sns_line = sns.lineplot(x='Time', y='value', data=df, hue='variable', style='variable', dashes=False,
                       markers=markers, ms=6, mec='k', mew=0.5, lw=2)
plt.xlabel('Time(d)')
plt.ylabel('value')
plt.xlim(-1, 20)
plt.ylim(-2, 90)
plt.xticks(np.linspace(0, 20, 11))
plt.yticks(np.linspace(0, 90, 10))
plt.legend(loc='best', frameon=False, title='variable');
```

```python
markers = ['o', 's', 'H', 'D']
fig = plt.figure(figsize=(6, 4), dpi=100)
sns_line = sns.lineplot(data=df, x='Time', y='value', hue='variable', style='variable', dashes=False,
                       markers=markers, ms=6, mec='k', mew=0.5, lw=2)
plt.xlabel('Time(d)')
plt.ylabel('value')
plt.xlim(-1, 20)
plt.ylim(-2, 90)
plt.xticks(np.linspace(0, 20, 11))
plt.yticks(np.linspace(0, 90, 10))
plt.legend(loc='best', frameon=False)
# sns.despine() # 去掉上部和右边边框

# 另一种方法去掉上和右边框
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
```

### 使用seaborn绘制带标记的分面曲线图

```python
sns.set_context("notebook", font_scale=1.5,
                rc={'font.size': 12,  
                    'axes.labelsize': 13, 'legend.fontsize':13, 
                    'xtick.labelsize': 12,'ytick.labelsize': 12})
g = sns.FacetGrid(df, col='variable', hue='variable', size=3, aspect=0.9, gridspec_kws={'wspace':0.1})
g.map(sns.lineplot, 'Time', 'value', marker='o', dashes=False, ms=6, mec='k', mew=0.5, lw=2)
g.set_xlabels('Time(d)')
g.set_ylabels('value')
plt.xticks(np.linspace(0, 20, 5))
plt.yticks(np.linspace(0, 80, 5))
g.set_titles(row_template='{row_name}', col_template='{col_name}');
```

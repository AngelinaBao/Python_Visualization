## matplotlib.pyplot.legend

**legend()<br>
legend(handles, labels)<br>
legend(handles=handles)<br>
legend(labels)**

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format='svg'
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))

ax.plot([1, 2, 3], label='Inline label')
ax.legend();
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
line, = ax.plot([1, 2, 3])
line.set_label('Label via method')
ax.legend();
```

**如果label是以下划线_开头或者是个空字符串，则不会生成到legend中，也就是说没有legend显示**

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
line1, = ax.plot([1, 2, 3])
line2, = ax.plot([2, 1, 3])
line3, = ax.plot([3, 1, 2])

ax.legend([line1, line2, line3], ['Line1', 'line2', 'line3']);
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
line1, = ax.plot([1, 2, 3], label='line1')
line2, = ax.plot([2, 1, 3], label='line2')
line3, = ax.plot([3, 1, 2], label='line3')

ax.legend(handles=[line1, line2, line3]);
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.plot([1, 2, 3])
ax.plot([5, 6, 7])
ax.legend(['First line', 'Second line'], frameon=False);
```

### Creating artists specifically for adding to the legend

#### Suppose we wanted to create a legend which has an entry for some data which is represented by a red color:

```python
import matplotlib.patches as mpatches

fig, ax = plt.subplots(figsize=(5, 2.7))
red_patch = mpatches.Patch(color='red', label='The red data')
ax.legend(handles=[red_patch]);
```

```python
import matplotlib.lines as mlines

fig, ax = plt.subplots(figsize=(5, 2.7))
blue_line = mlines.Line2D([], [], color='blue', marker='*',
                         markersize=15, label='Blue stars')
ax.legend(handles=[blue_line])
plt.show()
```

### Multiple legends on the same Axes

需要展现多个图例时，我们不能重复调用legend,这样只会有一个图例。为了保留旧的图例实例，我们需要手动把它加到axes.

```python
fig, ax = plt.subplots(figsize=(5, 3))
line1, = ax.plot([1, 2, 3], label='Line 1', linestyle='--')
line2, = ax.plot([3, 2, 1], label='Line2', linewidth=4)

# create a legend for line1

first_legend = ax.legend(handles=[line1], loc='upper right')

# add the legend manually to the axes
ax.add_artist(first_legend)

# create another legend for line2
ax.legend(handles=[line2], loc='lower right')

plt.show()
```

The following example demonstrates combining two legend keys on top of one another:

```python
from numpy.random import randn

z = randn(10)

fig, ax = plt.subplots()
red_dot, = ax.plot(z, 'ro', markersize=12)
# put a white cross over some of the data
white_cross, = ax.plot(z[:5], 'w+', markeredgewidth=3, markersize=12)

ax.legend([red_dot, (red_dot, white_cross)], ['Attr A', 'Attr A+B'])
```

The legend_handler.HandlerTuple class can also be used to assign several legend keys to the same entry:

```python
from matplotlib.legend_handler import HandlerLine2D, HandlerTuple

fig, ax = plt.subplots(figsize=(5, 2.7))
p1, = ax.plot([1, 2.5, 3], 'r-d')
p2, = ax.plot([3, 2, 1], 'k-o')

l = ax.legend([(p1, p2)], ['Two keys'], numpoints=1,
              handler_map={tuple: HandlerTuple(ndivide=None)})
```

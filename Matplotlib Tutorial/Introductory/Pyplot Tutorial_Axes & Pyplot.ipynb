### explicit "Axes" & implicit "Pyplot"

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='svg'
```

```python
fig = plt.figure()
ax = fig.subplots()
ax.plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
```

```python
plt.plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
```

```python
plt.subplot(121) # equivalent to plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [0, 0.5, 0.2])

plt.subplot(122)
plt.plot([3, 2, 1], [0, 0.5, 0.2]);
```

```python
plt.subplot(1, 2, 1) # equivalent to plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [0, 0.5, 0.2])

plt.subplot(1, 2, 2)
plt.plot([3, 2, 1], [0, 0.5, 0.2]);
```

```python
# axes
plt.subplot(1, 2, 1)
ax = plt.gca()
ax.plot([1, 2, 3], [0, 0.5, 0.2])

plt.subplot(1, 2, 2)
ax = plt.gca()
ax.plot([3, 2, 1], [0, 0.5, 0.2]);
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(121) 
ax1.plot([1, 2, 3], [0, 0.5, 0.2])
ax2 = fig.add_subplot(122)
ax2.plot([3, 2, 1], [0, 0.5, 0.2]);
```

```python
fig, ax = plt.subplots(1,  2)
ax[0].plot([1, 2, 3], [0, 0.5, 0.2])
ax[1].plot([3, 2, 1], [0, 0.5, 0.2]);
```

```python
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [0, 0.5, 0.2])

plt.subplot(1, 2, 2)
plt.plot([3, 2, 1], [0, 0.5, 0.2])

plt.suptitle('Implicit Interface: re-call subplot')

for i in range(1, 3):
    plt.subplot(1, 2, i)
    plt.xlabel('Boo')
```

```python
axs = []
ax = plt.subplot(1, 2, 1)
axs += [ax]
plt.plot([1, 2, 3], [0, 0.5, 0.2])

ax = plt.subplot(1, 2, 2)
axs += [ax]
plt.plot([3, 2, 1], [0, 0.5, 0.2])

plt.suptitle('Implicit Interface: save handles')

for i in range(2):
    plt.sca(axs[i])
    plt.xlabel('Boo')
```

However, the recommended way would be to be explicit from the outset:

```python
fig, ax = plt.subplots(1, 2)
ax[0].plot([1, 2, 3], [0, 0.5, 0.2])
ax[1].plot([3, 2, 1], [0, 0.5, 0.2])

plt.suptitle('Implicit Interface: save handles')

for i in range(2):
    ax[i].set_xlabel('Boo') # plt:xlabel, ax: set_xlabel
```

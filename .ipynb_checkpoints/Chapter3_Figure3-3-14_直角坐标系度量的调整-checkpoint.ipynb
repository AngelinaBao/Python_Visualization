```python
import pandas as pd
import numpy as np
import matplotlib as mpl
from plotnine import *
mpl.rcParams['font.family']='Times New Roman'
```

```python
df = pd.read_csv(r'C:\Users\1000257489\Documents\2022\Angelina\Study\Beautiful Visualization with Python\Beautiful-Visualization-with-python-master\第3章_数据可视化基础\MappingAnalysis_Data.csv')
df.shape
```

```python
df.head()
```

```python
p1 = (ggplot(df, aes(x='Time', y='value', fill='variable', shape='variable'))
      + geom_line()
      + geom_point(size=4, stroke=0.25, color='black')
      + scale_fill_manual(values=["#FF9641","#FF5B4E","#B887C3","#38C25D"])
      + theme_xkcd() 
      + theme(text = element_text(color='black', size=10), 
              aspect_ratio=0.8, dpi=100, figure_size=(5, 4),
              legend_position=(0.28, 0.72), legend_direction='vertical', 
              legend_background=element_blank())
     )
print(p1)
```

```python
p2 = (ggplot(df, aes(x='Time', y='value', fill='variable', shape='variable'))
      + geom_line()
      + geom_point(color='black', size=5, stroke=0.25)
      + scale_shape_manual(values=['o', 's', 'D', '^'])
      + scale_fill_manual(values=["#FF9641","#FF5B4E","#B887C3","#38C25D"])
      + scale_x_continuous(breaks=np.arange(0, 21, 2), name='Time(d)', limits=[0, 21])
      + scale_y_continuous(breaks=np.arange(0, 81, 10), name='value(M)', limits=[0, 81], expand=(0, 1))
      + labs(title='2020 Company Sales Comparison Plot')
      + theme_classic()
      + theme(text=element_text(color='black', size=10), dpi=100, figure_size=(5, 4),
             legend_position=(0.32, 0.7), legend_direction='vertical', legend_background=element_blank())
     )
print(p2)
```

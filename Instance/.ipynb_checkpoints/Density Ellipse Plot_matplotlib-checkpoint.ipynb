```python
def confidence_ellipse(x, y, ax, n_std=3.0, facecolor='none', **kwargs):
    """
    Create a plot of the covariance confidence ellipse of *x* and *y*.

    Parameters
    ----------
    x, y : array-like, shape (n, )
        Input data.

    ax : matplotlib.axes.Axes
        The axes object to draw the ellipse into.

    n_std : float
        The number of standard deviations to determine the ellipse's radiuses.

    **kwargs
        Forwarded to `~matplotlib.patches.Ellipse`

    Returns
    -------
    matplotlib.patches.Ellipse
    """
    if x.size != y.size:
        raise ValueError("x and y must be the same size")

    cov = np.cov(x, y)
    pearson = cov[0, 1]/np.sqrt(cov[0, 0] * cov[1, 1])
    # Using a special case to obtain the eigenvalues of this two-dimensional dataset.
    ell_radius_x = np.sqrt(1 + pearson)
    ell_radius_y = np.sqrt(1 - pearson)
    ellipse = Ellipse((0, 0), width=ell_radius_x * 2, height=ell_radius_y * 2,
                      facecolor=facecolor, **kwargs)

    # Calculating the standard deviation of x from the squareroot of the variance and multiplying 
    # with the given number of standard deviations.
    scale_x = np.sqrt(cov[0, 0]) * n_std
    mean_x = np.mean(x)

    # calculating the standard deviation of y ...
    scale_y = np.sqrt(cov[1, 1]) * n_std
    mean_y = np.mean(y)

    transf = transforms.Affine2D() \
        .rotate_deg(45) \
        .scale(scale_x, scale_y) \
        .translate(mean_x, mean_y)

    ellipse.set_transform(transf + ax.transData)
    return ax.add_patch(ellipse)
```

```python
df_new = df.sort_values(by=['Initial_SER'], ascending=True)
df_new.drop_duplicates(subset=['Initial_SER'], inplace=True)
df_new.shape
```

```python
df[['build', 'media', 'Track_group']].value_counts()
```

```python
df_1 = df.loc[df['media']=='SZ_SD4d_2']
df_2 = df.loc[df['media']=='SZ_SD5st']

df_11 = df_new.loc[df_new['media']=='SZ_SD4d_2']
df_22 = df_new.loc[df_new['media']=='SZ_SD5st']

fig, ax = plt.subplots(figsize=(3.5, 3.5))

p1 = confidence_ellipse(df_1['Initial_SER'], df_1['delSER_Max'], ax, n_std=2, edgecolor='#08519c', linestyle='dashed')
p2 = confidence_ellipse(df_2['Initial_SER'], df_2['delSER_Max'], ax, n_std=2, edgecolor='#9ecae1', linestyle='dashed')


tck, u = interpolate.splprep([df_11['Initial_SER'], df_11['delSER_Max']], s=0.1)
new_points = interpolate.splev(u, tck)
p11, = ax.plot(new_points[0], new_points[1], color='#08519c', linestyle='solid')

tck, u = interpolate.splprep([df_22['Initial_SER'], df_22['delSER_Max']], s=0.1)
new_points = interpolate.splev(u, tck)
p22, = ax.plot(new_points[0], new_points[1], color='#9ecae1', linestyle='solid')

ax.set_title('delSER_Max vs. Ini_SER')
ax.set_xlabel('Initial_SER')
ax.set_ylabel('delSER_Max')
ax.set_xlim(-2.1, -1.6)
ax.set_ylim(0.01, 0.16)
ax.legend([(p1, p11), (p2, p22)], ['SZ_SD4d_2', 'SZ_SD5st'], bbox_to_anchor=(1.05, 0.5))

plt.show()
```

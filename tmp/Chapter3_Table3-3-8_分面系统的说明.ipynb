```python
from plotnine import *
from plotnine.data import mpg
```

```python
mpg.head()
```

```python
mpg.fl.value_counts()
```

```python
t = (ggplot(mpg, aes(x='cty', y='hwy', fill='fl'))
     + geom_point(size=3, stroke=0.3, alpha=0.8, show_legend=False)
     + scale_fill_hue(s=0.9, l=0.65, h=0.0417, color_space='husl')
     + theme(figure_size=(4, 3), dpi=100)
    )
print(t)
```

```python
mpg['fl'].value_counts()
```

```python
t1 = (t + facet_grid('. ~ fl') # .表示在这个维度不用分面，这个语句仅在列方向分面，行方向不用分
      + theme_matplotlib()
      + theme(text=element_text(color='black', size=10), 
             strip_text=element_text(color='black', size=15), # 设置分面标签的字体大小和颜色
             strip_background=element_rect(color='black', fill='lightgray'), #设置分面标签的外框颜色和填充颜色
             aspect_ratio=1.3,
             dpi=100,
             figure_size=(10, 5)) # 设置不同的figure_size,可以改变facet灰色框部分的大小
     )
print(t1)
```

```python
t2 = (t + facet_grid('year ~ .')
      + theme_xkcd()
      + theme(text=element_text(color='black', size=12),
             strip_text=element_text(color='black', size=12), # theme里的text设置都需要借助于element_text
             strip_background=element_rect(color='black'),
             aspect_ratio=0.5,
             dpi=100,
             figure_size=(4, 4)) 
     )
print(t2)
```

```python
t3 = (t + facet_grid('year ~ fl')
      + theme_matplotlib()
      + theme(text=element_text(color='black', size=12),
             strip_text=element_text(color='black', size=15), # theme里的text设置都需要借助于element_text
             strip_background=element_rect(color='black'),
             aspect_ratio=1,
             dpi=100,
             figure_size=(10, 8)) 
     )
print(t3)
```

```python
t4 = (t + facet_wrap('~ fl') # t1利用facdet_grid按列排，facet_wrap按矩形排，会换行
      + theme_matplotlib()
      + theme(text=element_text(color='black', size=12),
             strip_text=element_text(color='black', size=15),
             strip_background=element_rect(color='black'),
             aspect_ratio=1,
             dpi=100,
             figure_size=(8, 5)) 
     )
print(t4)
```

```python
t5 = (t + facet_grid('drv ~ fl', scales='fixed') 
      # scales:调整x轴和y轴的取值范围, fixed:对所有的subgroup,x和y轴取相同范围;free:x,y轴根据自己的取值选取范围;free_x:x轴自由选取范围，y轴保持相同
      + theme_matplotlib()
      + theme(text=element_text(color='black', size=12),
             strip_text=element_text(color='black', size=15),
             strip_background=element_rect(color='black'),
             aspect_ratio=1,
             dpi=100,
             figure_size=(8, 5)) 
     )
print(t5)
```

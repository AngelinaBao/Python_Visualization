```python
import pandas as pd
import numpy as np
from plotnine import *
```

```python
df = pd.read_csv(r'C:\Users\1000257489\Documents\2022\Angelina\Study\Beautiful Visualization with Python\Beautiful-Visualization-with-python-master\第3章_数据可视化基础\Bubble_Data.csv')
df.shape
```

```python
df.head()
```

```python
p1 = (ggplot(df, aes(x='Gas Mileage(mpg)', y='Power (kW)'))
      + geom_point(shape='o', fill='orange', color='black', size=8, stroke=0.2)
      + theme_light()
      + theme(
#           text=element_text(color='black', size=10), dpi=100, figure_size=(5, 4)
          axis_title = element_text(size=16, face='plain', color='black'),
          axis_text = element_text(size=14, face='plain', color='black'),
          legend_text = element_text(size=14, face='plain', color='black'),
          legend_title = element_text(size=16, face='plain', color='black'),
          legend_background=element_blank(),
          legend_position=(0.81, 0.75),
          figure_size=(8, 8),
          dpi=50            
             )
     )
print(p1)
```

```python
p2 = (ggplot(df, aes(x='Gas Mileage(mpg)', y='Power (kW)', size='Weight (kg)'))
      + geom_point(shape='o', fill='orange', color='black', stroke=0.2)
      + scale_size_continuous(range=[1, 15])
      + theme_light()
      + theme(
#           text=element_text(color='black', size=10), dpi=100, figure_size=(5, 4)
          axis_title = element_text(size=16, face='plain', color='black'),
          axis_text = element_text(size=14, face='plain', color='black'),
          legend_text = element_text(size=14, face='plain', color='black'),
          legend_title = element_text(size=16, face='plain', color='black'),
          legend_background=element_blank(),
          legend_position=(0.81, 0.75),
          legend_direction='vertical',
          figure_size=(8, 8),
          dpi=50)
     )
print(p2)
```

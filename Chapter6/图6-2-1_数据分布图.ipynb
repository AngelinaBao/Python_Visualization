```python
from pandas.api.types import CategoricalDtype
cat_dtype = CategoricalDtype(categories=['n','s','k','mm'], ordered=True)
df['class'] = df['class'].astype(cat_dtype)
```

#### 核密度估计图

```python
density_plot = (ggplot(df, aes(x='value', fill='class'))
                + geom_density(alpha=0.8, size=0.2)
                + facet_grid('class~.')
                + scale_fill_hue(s=0.9, l=0.65, h=0.04, color_space='husl')
                + theme_light()
                + theme(axis_title=element_text(size=15, color='k', face='plain'),
                        axis_text=element_text(size=12, color='k', face='plain'),
                        legend_title=element_text(size=12, color='k', face='plain'),
                        legend_text=element_text(size=10, color='k', face='plain'),
                        strip_text=element_text(color='k', size=14),
                        strip_background=element_rect(fill='lightgrey'),
                        aspect_ratio=0.25,
                        dpi=100,
                        figure_size=(4, 4))
               )
print(density_plot)
```

```python
density_plot.save('density_plot.png', dpi=200)
```

#### 统计直方图

```python
histogram_plot = (ggplot(df, aes(x='value', fill='class'))
                + geom_histogram(color='k', bins=30, size=0.2, alpha=0.8)
                + facet_grid('class~.')
                + scale_fill_hue(s=0.9, l=0.65, h=0.04, color_space='husl')
                + theme_light()
                + theme(axis_title=element_text(size=15, color='k', face='plain'),
                        axis_text=element_text(size=12, color='k', face='plain'),
                        legend_title=element_text(size=12, color='k', face='plain'),
                        legend_text=element_text(size=10, color='k', face='plain'),
                        strip_text=element_text(color='k', size=14),
                        strip_background=element_rect(fill='lightgrey'),
                        aspect_ratio=0.25,
                        dpi=100,
                        figure_size=(4, 4))
               )
print(histogram_plot)
```

```python
histogram_plot.save('histogram_plot.png', dpi=200)
```

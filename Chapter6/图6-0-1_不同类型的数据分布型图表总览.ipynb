```python
import pandas as pd
import numpy as np
from plotnine import *
```

```python
data = pd.read_csv(r'C:\Users\1000257489\Documents\2022\Angelina\Study\Beautiful_Visualization_with_Python\Beautiful-Visualization-with-python-master\第6章 数据分布型图表\Norm.csv')
data.shape
```

```python
data.head()
```

```python
data['Class'] = 'Class'
data.shape, data.head()
```

```python
data.Value.describe()
```

### 散点抖动图

```python
p1 = (ggplot(data, aes('Class', 'Value'))
      + geom_jitter(fill='#a6d854', position=position_jitter(0.2), shape='o', size=4, stroke=0.2)
      + scale_y_continuous(breaks=np.arange(0, 7, 1))
      + theme_matplotlib()
      + theme(panel_background=element_rect(fill="white",colour="black",size=0.25),
              axis_line=element_line(colour="black",size=0.25),
              axis_title=element_text(size=13,face="plain",color="black"),
              axis_text = element_text(size=12,face="plain",color="black"),
              axis_title_x = element_blank(),
              legend_position='none',
#               aspect_ratio=1.5,
              figure_size=(3, 5),
              dpi=100
             )
     )
print(p1)
```

### 点阵图

```python
p2 = (ggplot(data, aes('Class', 'Value'))
      + geom_dotplot(fill='#a6d854', binaxis='y', stackdir='center', dotsize=0.3)
      + scale_y_continuous(breaks=np.arange(0, 7, 1))
      + theme_matplotlib()
      + theme(panel_background=element_rect(fill="white",colour="black",size=0.25),
              axis_line=element_line(colour="black",size=0.25),
              axis_title=element_text(size=13,face="plain",color="black"),
              axis_text = element_text(size=12,face="plain",color="black"),
              axis_title_x = element_blank(),
              legend_position='none',
              aspect_ratio=1.5,
              figure_size=(3, 3),
              dpi=100)
     )
print(p2)
```

### 统计直方图

```python
p3 = (ggplot(data, aes('Value', fill='Value'))
      + geom_histogram(fill='#a6d854', color='black', size=0.2, bins=20)
      + coord_flip()
      + scale_x_continuous(breaks=np.arange(0, 7, 1))
      + theme_matplotlib()
      + theme(panel_background=element_rect(color='black'),
              text=element_text(size=15, color='black'),
              plot_title=element_text(size=15,family="myfont",face="bold.italic",hjust=.5,color="black"),
              legend_position=(0.8, 0.8),
              legend_background=element_blank(),
              figure_size=(3, 5)
             )
     )
print(p3)
```

### 核密度估计图

```python
p4 = (ggplot(data, aes('Value', fill='Value'))
      + geom_density(fill='#a6d854', color='black', size=0.25, alpha=0.8)
      + coord_flip()
      + scale_x_continuous(breaks=np.arange(0, 7, 1))
      + theme_matplotlib()
      + theme(panel_background=element_rect(color='black'),
              text=element_text(size=15, color='black'),
              plot_title=element_text(size=15,family="myfont",face="bold.italic",hjust=.5,color="black"),
              legend_position=(0.8, 0.8),
              legend_background=element_blank(),
              figure_size=(3, 5)
             )
     )
print(p4)
```

### 箱型图

```python
p5 = (ggplot(data, aes('Class', 'Value'))
      + geom_boxplot(fill='#a6d854', notch=False) # notch是否有凹痕
      + scale_y_continuous(breaks=np.arange(0, 7, 1))
      + theme_matplotlib()
      + theme(panel_background=element_rect(fill='white', color='black', size=0.2),
             axis_line=element_line(colour="black",size=0.25),
             axis_title=element_text(size=13,face="plain",color="black"),
             axis_text = element_text(size=12,face="plain",color="black"),
             legend_position="none",
             figure_size=(3, 5))
     )
print(p5)
```

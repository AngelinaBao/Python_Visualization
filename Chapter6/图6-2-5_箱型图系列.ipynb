```python
import pandas as pd
from plotnine import *
```

```python
df = pd.read_csv('Distribution_Data.csv')
df.shape
```

```python
from pandas.api.types import CategoricalDtype
cat_dtype = CategoricalDtype(categories=df['class'].unique(), ordered=True)
df['class'] = df['class'].astype(cat_dtype)
```

```python
box_plot = (ggplot(df, aes(x='class', y='value', fill='class'))
            + geom_boxplot(show_legend=False)
            + scale_fill_hue(s=0.9, l=0.65, h=0.04, color_space='husl')
            + theme_matplotlib()
            + theme(axis_title=element_text(size=15, color='k', face='plain'),
                   axis_text=element_text(size=12, color='k', face='plain'),
                   legend_title=element_text(size=12, color='k', face='plain'),
                   legend_text=element_text(size=12, color='k', face='plain'),
                   aspect_ratio=1.15,
                   dpi=100,
                   figure_size=(4, 4))
           )
print(box_plot)
```

```python
box_jitter_plot = (ggplot(df, aes(x='class', y='value', fill='class'))
            + geom_boxplot(show_legend=False)
            + geom_jitter(fill='black', shape='.', width=0.3, size=2, stroke=0.1, alpha=0.7, show_legend=False)
            + scale_fill_hue(s=0.9, l=0.65, h=0.04, color_space='husl')
            + theme_matplotlib()
            + theme(axis_title=element_text(size=15, color='k', face='plain'),
                   axis_text=element_text(size=12, color='k', face='plain'),
                   legend_title=element_text(size=12, color='k', face='plain'),
                   legend_text=element_text(size=12, color='k', face='plain'),
                   aspect_ratio=1.15,
                   dpi=100,
                   figure_size=(4, 4))
           )
print(box_jitter_plot)
```

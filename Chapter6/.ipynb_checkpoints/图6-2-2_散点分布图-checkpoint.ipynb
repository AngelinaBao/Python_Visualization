```python
import pandas as pd
import numpy as np
from plotnine import *
```

```python
df = pd.read_csv('Distribution_Data.csv')
df.shape
```

```python
df.head()
```

```python
df['class'].unique()
```

```python
from pandas.api.types import CategoricalDtype
cat_dtype = CategoricalDtype(categories=df['class'].unique(), ordered=True)
df['class'] = df['class'].astype(cat_dtype)
```

#### 抖动散点图

```python
jitter_plot = (ggplot(df, aes(x='class', y='value', fill='class'))
               + geom_jitter(width=0.3, size=3, stroke=0.1, show_legend=False)
               + scale_fill_hue(s=0.9, l=0.65, h=0.04, color_space='husl')
               + theme_matplotlib()
               + theme(axis_title=element_text(color='k', size=15, face='plain'),
                      axis_text=element_text(color='k', size=12, face='plain'),
                      legend_title=element_text(color='k', size=12, face='plain'),
                      legend_text=element_text(color='k', size=10, face='plain'),
                      aspect_ratio=1.15,
                      dpi=100,
                      figure_size=(4, 4))
              )
print(jitter_plot)
```

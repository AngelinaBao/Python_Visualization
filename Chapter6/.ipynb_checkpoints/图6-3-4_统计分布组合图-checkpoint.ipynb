```python
import pandas as pd
import numpy as np
from plotnine import *
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('ticks')
sns.set_context("notebook", font_scale=1.5,
                rc={'font.size': 13, 
                    'axes.labelsize': 16, 
                    'legend.fontsize':16, 
                    'xtick.labelsize': 15,
                    'ytick.labelsize': 15})
```

```python
# 二维散点与统计直方图
sns_hist = sns.jointplot(x='total_bill', y='tip',
                        data=tips,
                        color = '#D31A8A',
                        s = 60, edgecolor="w",linewidth=0.5,# 设置散点大小、边缘线颜色及宽度(只针对scatter）
                        kind = 'scatter', #kind="kde","hex","reg"
                        space = 0,  # 设置散点图和布局图的间距
                        height = 6, ratio = 6,  # 散点图与布局图高度比，整型
                        xlim=(0,60),
                        marginal_kws=dict(bins=20, 
                                          edgecolor='k',
                                          color='#7CBC47')  # 设置柱状图箱数，是否设置rug
              )

sns_hist.set_axis_labels(xlabel='total bill', ylabel='tip');
```

```python
# 带趋势线的二维散点与统计直方图
sns_reg = sns.jointplot(x='total_bill', y='tip',
                        data=tips,
                        color = '#D31A8A',
                        kind = 'reg', #kind="kde","hex","reg"
                        space = 0,  # 设置散点图和布局图的间距
                        height = 6, ratio = 6,  # 散点图与布局图高度比，整型
                        line_kws={"color":"#D31A8A","alpha":1,"lw":4},
                        scatter_kws={"color":"#7CBC47","alpha":0.7,"s":30,'marker':"+"},
                        marginal_kws=dict(bins=20, 
                                          edgecolor='k',
                                          color='#7CBC47')  # 设置柱状图箱数，是否设置rug
              )

sns_reg.set_axis_labels(xlabel='total bill', ylabel='tip');
```

```python
# 二维与一维统计直方图
sns_hex = sns.jointplot(x='total_bill', y='tip',
                        data=tips,
                        color = '#D31A8A', linewidth=0.1,
                        kind = 'hex', #kind="kde","hex","reg"
                        space = 0,  # 设置散点图和布局图的间距
                        height = 6, ratio = 6,  # 散点图与布局图高度比，整型
                        xlim=(0,60),
                        joint_kws=dict(gridsize=20,edgecolor='w'),
                        marginal_kws=dict(bins=20, 
                                          edgecolor='k',
                                          color='#D31A8A') 
              )

sns_hex.set_axis_labels(xlabel='total bill', ylabel='tip');
```

```python
# 二维与一维核密度估计图
sns_kde =sns.jointplot(x='total_bill', y='tip', data=tips, kind="kde",color='#D31A8A')
sns_kde.plot_joint(plt.scatter, c="k", s=10, linewidth=1, marker="+") 
sns_kde.set_axis_labels(xlabel='total bill', ylabel='tip');
```

```python
sns_kde =sns.jointplot(x='total_bill', y='tip', data=tips, kind="kde",color='#D31A8A', fill='..density..')
sns_kde.plot_joint(plt.scatter, c="k", s=10, linewidth=1, marker=".", alpha=0.4) 
sns_kde.set_axis_labels(xlabel='total bill', ylabel='tip');
```

```python
sns_kde =sns.jointplot(x='total_bill', y='tip', data=tips, kind="kde",color='#D31A8A', fill='..level..', levels=100)
sns_kde.plot_joint(plt.scatter, c="k", s=10, linewidth=1, marker=".", alpha=0.4) 
sns_kde.set_axis_labels(xlabel='total bill', ylabel='tip');
```

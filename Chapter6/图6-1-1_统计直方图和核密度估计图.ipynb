```python
import pandas as pd
from plotnine import *
```

```python
df = pd.read_csv(r'C:\Users\1000257489\Documents\2022\Angelina\Study\Beautiful_Visualization_with_Python\Beautiful-Visualization-with-python-master\第6章 数据分布型图表\Hist_Density_Data.csv')
df.shape
```

```python
df.head()
```

#### 单数据系列统计直方图

```python
p3 = (ggplot(df, aes(x='MXSPD'))
      + geom_histogram(binwidth=1, color='k', alpha=1, size=0.25, fill='#fb9a99')
      + scale_fill_hue(s=0.9, l=0.65, h=0.015, color_space='husl')
      + theme(
          text=element_text(size=13,color="black"),
          plot_title=element_text(size=15,family="myfont",face="bold.italic",hjust=.5,color="black"),#,
          legend_position='none',
          aspect_ratio =1.15,
          figure_size=(5,5)
      )
     )
print(p3)
```

#### 单数据系列核密度估计图

```python
p4 = (ggplot(df, aes(x='MXSPD'))
      + geom_density(bw=1, color='k', alpha=1, size=0.25, fill='#fb9a99', kernel='Gaussian')
      + scale_fill_hue(s=0.9, l=0.65, h=0.015, color_space='husl')
      + theme(
          text=element_text(size=13,color="black"),
          plot_title=element_text(size=15,family="myfont",face="bold.italic",hjust=.5,color="black"),#,
          legend_position='none',
          aspect_ratio =1.15,
          figure_size=(5,5)
      )
     )
print(p4)
```

#### 多数据系列直方图

```python
p1 = (ggplot(df, aes(x='MXSPD', fill='Location'))
      + geom_histogram(binwidth=1, color='k', alpha=0.6, size=0.25)
      + scale_fill_hue(s=0.9, l=0.65, h=0.015, color_space='husl')
      + theme(
          text=element_text(size=13,color="black"),
          plot_title=element_text(size=15,family="myfont",face="bold.italic",hjust=.5,color="black"),#,
          legend_position=(0.7,0.8),
          legend_direction='vertical',
          legend_background = element_blank(),
          aspect_ratio =1.15,
          figure_size=(5,5)
      )
     )
print(p1)
```

#### 多数据系列核密度估计图

```python
p2 = (ggplot(df, aes(x='MXSPD', fill='Location'))
      + geom_density(bw=1,alpha=0.55,colour="black",size=0.25,kernel="gaussian")
      + scale_fill_hue(s=0.9, l=0.65, h=0.015, color_space='husl')
      + theme(
          text=element_text(size=13,color="black"),
          plot_title=element_text(size=15,family="myfont",face="bold.italic",hjust=.5,color="black"),#,
          legend_position=(0.7,0.8),
          legend_direction='vertical',
          legend_background = element_blank(),
          aspect_ratio =1.15,
          figure_size=(5,5)
      )
     )
print(p2)
```

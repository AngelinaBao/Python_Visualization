```python
import pandas as pd
import numpy as np
from plotnine import *
```

```python
df = pd.read_csv(r'C:\Users\1000257489\Documents\2022\Angelina\Study\Beautiful Visualization with Python\Beautiful-Visualization-with-python-master\第3章_数据可视化基础\MappingAnalysis_Data.csv')
df.shape
```

```python
df.head()
```

```python
p1 = (ggplot(df, aes(x='Time', y='value', fill='variable', shape='variable'))
      + geom_line(color='black')
      + geom_point(color='black', size=5, stroke=0.25, alpha=1)
      + scale_x_continuous(breaks=np.arange(0, 21, 5), name='Time(d)')
      + scale_y_continuous(breaks=np.arange(0, 81, 10))
      + theme_classic()
      + theme(figure_size=(4, 4), legend_background=element_rect(fill='white'), legend_position='right')
     )
print(p1)
```

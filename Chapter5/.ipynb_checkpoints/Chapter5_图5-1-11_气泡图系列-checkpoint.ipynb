```python
import pandas as pd
import numpy as np
from plotnine import *
from plotnine.data import mtcars
```

```python
mtcars.head()
```

```python
p1 = (ggplot(mtcars, aes(x='wt', y='mpg', size='disp'))
      + geom_point(shape='o', fill='#2b8cbe', color='black', stroke=0.2, alpha=0.8)
#       + guides(fill=guide_legend(title='disp'))
#       + theme_light()
      + theme(axis_title=element_text(size=16, color='k', face='plain'),
             axis_text=element_text(size=12, color='k', face='plain'),
             aspect_ratio=1,
             figure_size=(4, 4),
             dpi=100)
     )
print(p1)
```

```python
p2 = (ggplot(mtcars, aes(x='wt', y='mpg', size='disp', fill='disp'))
      + geom_point(shape='o', color='black', stroke=0.2, alpha=0.8)
      + guides(fill=guide_legend(title='disp'), size=guide_legend(title='disp'))    
      + theme(axis_title=element_text(size=16, color='k', face='plain'),
             axis_text=element_text(size=12, color='k', face='plain'),
             aspect_ratio=1,
             figure_size=(4, 4),
             dpi=100)
     )
print(p2)
```

```python
p3 = (ggplot(mtcars, aes(x='wt', y='mpg', size='disp', fill='disp'))
      + geom_point(shape='o', color='black', stroke=0.2, alpha=0.8)
      + scale_fill_gradient2(low="#377EB8",high="#E41A1C",
                             limits = (0,np.max(mtcars.disp)), 
                             midpoint = np.mean(mtcars.disp)) #(Colormap)
      + scale_size_area(max_size=12) #  
      + geom_text(label = mtcars.disp,size=8, nudge_x =0.3,nudge_y =0.6) # disp
      + theme(axis_title=element_text(size=16, color='k', face='plain'),
             axis_text=element_text(size=12, color='k', face='plain'),
             aspect_ratio=1,
             figure_size=(4, 4),
             dpi=100)
     )
print(p3)
```

```python
p4 = (ggplot(mtcars, aes(x='wt', y='mpg', size='disp', fill='disp'))
      + geom_point(shape='s', color='black', stroke=0.2, alpha=0.8)
      + scale_fill_gradient2(low="#377EB8",high="#E41A1C",
                             limits = (0,np.max(mtcars.disp)), 
                             midpoint = np.mean(mtcars.disp)) #(Colormap)
      + scale_size_area(max_size=12) #  
#       + guides(fill=guide_legend(title='disp'), size=guide_legend(title='disp'))    
      + theme(axis_title=element_text(size=16, color='k', face='plain'),
             axis_text=element_text(size=12, color='k', face='plain'),
             aspect_ratio=1,
             figure_size=(4, 4),
             dpi=100)
     )
print(p4)
```

```python
import pandas as pd
import numpy as np
from plotnine import *
```

```python
df = pd.DataFrame(dict(x=np.random.normal(loc=0, scale=1, size=250)))

base_plot = (ggplot(df, aes(sample='x')) # sample: sample quantiles
             + geom_qq(shape='o', fill='none')
             + geom_qq_line()
             + theme_classic()
             + theme(axis_title=element_text(size=18, color='black', face='plain'),
                    axis_text=element_text(size=16, color='black', face='plain'),
                    aspect_ratio=1.05,
                    figure_size=(4, 4),
                    dpi=100)
            )
print(base_plot)
```

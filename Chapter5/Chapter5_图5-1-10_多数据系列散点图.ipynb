```python
import pandas as pd
from plotnine import *
```

```python
df = pd.read_csv(r'C:\Users\1000257489\Documents\2022\Angelina\Study\Beautiful_Visualization_with_Python\Beautiful-Visualization-with-python-master\第5章 数据关系型图表\MultiSeries_Scatter_Data.csv')
df.shape, df.head()
```

```python
p1 = (ggplot(df, aes(x='x', y='y', color='factor(label_pred)', fill='factor(label_pred)'))
      + geom_point(shape='o', size=4, alpha=0.8, color='k', stroke=0.2)
      + scale_fill_manual(values=["#00AFBB","#FC4E07"])
#       + labs(x='Axis X', y='Axis y') 
      + scale_x_continuous(name='Axis X', limits=[-5, 10])
      + scale_y_continuous(name='Axis Y', limits=[-5, 10])
      + guides(fill=guide_legend(title='Cluster'), color=guide_legend(title='Cluster'))
      + theme(axis_title=element_text(size=16, color='k', face='plain'),
             axis_text=element_text(size=12, color='k', face='plain'),
             legend_position=(0.8, 0.2),
             legend_direction='vertical',
             legend_background=element_rect(fill='none'),
             aspect_ratio=1,
             figure_size=(5, 5),
             dpi=100)
     )
print(p1)
```

```python
p2 = (ggplot(df, aes(x='x', y='y', shape='factor(label_pred)', fill='factor(label_pred)'))
      + geom_point(size=4, alpha=0.8, color='k', stroke=0.1)
      + scale_fill_manual(values=["#00AFBB","#FC4E07"])
      + scale_color_manual(values=["#00AFBB","#FC4E07"])
      + scale_shape_manual(values=['o', 's'])
      + scale_x_continuous(name='Axis X', limits=[-5, 10])
      + scale_y_continuous(name='Axis Y', limits=[-5, 10])
      + guides(fill=guide_legend(title='Cluster'), shape=guide_legend(title='Cluster'))
      + theme(axis_title=element_text(size=16, color='k', face='plain'),
             axis_text=element_text(size=12, color='k', face='plain'),
             legend_title=element_text(size=12, color='k', face='plain'),
             legend_text=element_text(size=10, color='k', face='plain'),
             legend_position=(0.8, 0.2),
             legend_direction='vertical',
             legend_background=element_blank(),
             aspect_ratio=1,
             figure_size=(5, 5),
             dpi=100)
     )
print(p2)
```

```python
p2.save('多数据系列散点图.png', dpi=200)
```

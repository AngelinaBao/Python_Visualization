```python
import pandas as pd
import numpy as np
from plotnine import *
from datetime import datetime
```

```python
df=pd.read_csv('Line_Data.csv')

df['date']=[datetime.strptime(d, '%Y/%m/%d').date() for d in df['date']]

df.shape, df.head()
```

```python
df_melt = pd.melt(df, id_vars='date', var_name='variable', value_name='value')
df_melt.head()
```

```python
# 折线图
base_plot = (ggplot(df_melt, aes(x='date', y='value', group='variable', color='variable'))
             + geom_line(size=1)
             + scale_x_date(date_labels = "%Y",date_breaks = "2 year")
             + scale_fill_hue(s = 0.90, l = 0.65, h=0.0417,color_space='husl')
             + xlab('Year')
             + ylab('Value')
             + theme_matplotlib()
             + theme(axis_title=element_text(size=12, color='k', face='plain'),
                    axis_text=element_text(size=10, color='k', face='plain'),
                    legend_position=(0.25, 0.75),
                    legend_direction='vertical',
                    legend_background=element_blank(),
                    aspect_ratio=0.85,
                    dpi=100,
                    figure_size=(5, 5))
            )
print(base_plot)
```

```python
# 面积图
area_plot = (ggplot(df_melt, aes(x='date', y='value', group='variable'))
             + geom_area(aes(fill='variable'), alpha=0.75, position='identity')
             + geom_line(aes(color='variable'), size=0.75)
             + scale_x_date(date_labels='%Y', date_breaks='2 years')
             + scale_fill_hue(s = 0.90, l = 0.65, h=0.0417,color_space='husl')
             + xlab('Year')
             + ylab('Value')
             + theme_matplotlib()
             + theme(axis_title=element_text(size=12, color='k', face='plain'),
                    axis_text=element_text(size=12, color='k', face='plain'),
                    legend_position=(0.25, 0.75),
                    legend_direction='vertical',
                    legend_background=element_blank(),
                    aspect_ratio=0.85,
                    dpi=100,
                    figure_size=(5, 5))
            )
print(area_plot)
```

```python
# 面积图 纯色填充
df=pd.read_csv('Area_Data.csv')

df['date']=[datetime.strptime(d, '%Y/%m/%d').date() for d in df['date']]#df['date'].map(lambda x:datetime.datetime.strptime(x, '%Y/%m/%d').date())


base_plot=(ggplot(df, aes(x ='date', y = 'value',group=1) )
           + geom_area(fill="#FF6B5E",alpha=0.75,color='none') 
           + geom_line(color="black",size=0.75)
           + scale_x_date(date_labels = "%Y",date_breaks = "2 year")
           + xlab("Year")
           + ylab("Value")
           + theme(axis_title=element_text(size=10,face="plain",color="black"),
                   axis_text = element_text(size=10,face="plain",color="black"),
                   aspect_ratio =0.85,
                   figure_size = (5, 5),
                   dpi = 100)
          )
  
print(base_plot)
```

```python
import pandas as pd
import numpy as np
from plotnine import *
import matplotlib as mpl
# import matplotlib.pyplot as plt
mpl.rcParams['font.family'] = 'Times New Roman'
```

图a: 位置+方向

```python
p1 = (ggplot(df, aes(x='Time', y='value', group='variable'))
      + geom_line()   # 若先执行geom_point, line会穿过point
      + geom_point(shape='o', fill='white', size=3, stroke=0.25) 
      + theme_classic()
      + theme(text=element_text(color='black', size=12), 
              aspect_ratio=0.8, 
              dpi=100, figure_size=(4, 4),
              legend_position=(0.32, 0.75)
             )
     )
print(p1)
```

```python
p2 = (ggplot(df, aes(x='Time', y='value', fill='variable'))
      + geom_line()
      + geom_point(shape='o', size=3, stroke=0.25)
      + scale_fill_manual(values=["#595959","#BFBFBF","black","white"])
      + theme_classic()
      + theme(text=element_text(color='black', size=10), 
              aspect_ratio=0.8, 
              dpi=100, 
              figure_size=(4, 4), 
              legend_position=(0.3, 0.7),
              legend_direction='vertical',
              legend_background=element_rect(fill="none"))
      + guides(ncol=1)
     )
print(p2)
```

```python
p3 = (ggplot(df, aes(x='Time', y='value', shape='variable'))
      + geom_line()
      + geom_point(size=3, fill="#BFBFBF", stroke=0.25)
      + scale_shape_manual(values=['o', 's', 'D', '^'])
      + theme_classic()
      + theme(text=element_text(color='black', size=10), 
              aspect_ratio=0.8, 
              dpi=100, 
              figure_size=(4,4),
              legend_position=(0.32, 0.7),
              legend_direction='vertical',
              legend_background=element_rect(fill='none')) # element_rect is Used for backgrounds and borders
     )
print(p3)
```

```python
p4 = (ggplot(df, aes(x='Time', y='value', fill='variable'))
      + geom_line() # 如果先使用geom_point, 则线会穿过点
      + geom_point(shape='o', color='black', size=3, stroke=0.25)
      + scale_fill_manual(values=["#FF9641","#FF5B4E","#B887C3","#38C25D"])
      + theme_classic()
      + theme(text=element_text(color='black', size=10), 
             aspect_ratio=0.8,
             dpi=100,
             figure_size=(4, 4),
             legend_position=(0.3, 0.7),
             legend_direction='vertical',
             legend_margin=False,
             legend_background=element_rect(fill='None'))  # element_rect is Used for backgrounds and borders
      + labs(x='Time(s)', y='Value(M)', title='2022 Sales Trend per Group') # add labels for x and y axis
     )
print(p4)
```

```python
p5 = (ggplot(df, aes(x='Time', y='value', shape='variable', fill='variable'))
      + geom_line()
      + geom_point(size=3, color='black', stroke=0.25)
      + scale_fill_manual(values=["#595959","#BFBFBF","black","white"])
      + theme_classic()
      + theme(text=element_text(color='black', size=10),
             aspect_ratio=0.8,
             dpi=100,
             figure_size=(4, 4),
             legend_position=(0.3, 0.7),
             legend_direction='vertical',
             legend_background=element_rect(fill='None'))
     )

print(p5)
```

```python
p5 = (ggplot(df, aes(x='Time', y='value', shape='variable', fill='variable'))
      + geom_line()
      + geom_point(size=3, color='black', stroke=0.25) # stroke设置点的edge宽度
      + scale_shape_manual(values=['o', 's', 'D', '^'])
      + scale_fill_manual(values=["#FF9641","#FF5B4E","#B887C3","#38C25D"])
      + scale_x_continuous(name='Time(s)', breaks=np.arange(0, 21, 2), limits=(0, 20)) # 设置x轴ticks和label
      + scale_y_continuous(breaks=np.arange(0, 91, 10), limits=(0, 90), expand=(0, 1))
      + theme_classic()
      + theme(text=element_text(color='black', size=10),
             aspect_ratio=0.8,
             dpi=100,
             figure_size=(4, 4),
             legend_position=(0.3, 0.7),
             legend_direction='vertical',
             legend_background=element_rect(fill='None'))
      + labs(y='Value(M)', title='2020 Company Sales Trend Analysis')
     )

print(p5)
```
